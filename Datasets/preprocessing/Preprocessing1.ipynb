```python
df_maha = pd.read_csv("maha_results_2019.csv")
df_maha
```

```python
# Initialize empty columns for turnout data
df_maha['GENERAL_TURNOUT'] = None
df_maha['POSTAL_TURNOUT'] = None
df_maha['total_votes_in_state'] = None
df_maha['PERCENT_VOTES_POLLED'] = None

# Loop through the DataFrame to identify and process turnout rows
for i, row in df_maha.iterrows():
    if row[0] == ' TURNOUT ':  # Identify the turnout row by the "TURNOUT" label
        # Extract turnout data from this row
        general_turnout = df_maha.iloc[i, df_maha.columns.get_loc('general_votes')]
        postal_turnout = df_maha.iloc[i, df_maha.columns.get_loc('postal_votes')]
        total_votes_polled = df_maha.iloc[i, df_maha.columns.get_loc('secured_votes')]
        percent_votes_polled = df_maha.iloc[i, df_maha.columns.get_loc('percentage_of_votes')]
        
        # Find the previous rows for the same constituency and fill turnout columns
        for j in range(i - 1, -1, -1):
            if df_maha.iloc[j, df_maha.columns.get_loc('state_name')] == df_maha.iloc[i - 1, df_maha.columns.get_loc('state_name')]:
                df_maha.at[j, 'GENERAL_TURNOUT'] = general_turnout
                df_maha.at[j, 'POSTAL_TURNOUT'] = postal_turnout
                df_maha.at[j, 'total_votes_in_state'] = total_votes_polled
                df_maha.at[j, 'PERCENT_VOTES_POLLED'] = percent_votes_polled
            else:
                break

# Remove turnout rows to clean the dataset
df = df_maha[df_maha['state_name'] != ' TURNOUT ']

# Save the modified DataFrame to a new CSV
df.to_csv('updated_maha_dataset.csv', index=False)

print("Turnout data integrated and CSV exported.")
```

```python
# Load the dataset
df_new = pd.read_csv('updated_maha_dataset.csv')

# Create a new column 'candidate_position' and extract the position number from 'candidate_name'
df_new['candidate_position'] = df_new['candidate_name'].str.extract(r'^(\d+)', expand=False)

# Remove the position number from the 'candidate_name' column
df_new['candidate_name'] = df_new['candidate_name'].str.replace(r'^\d+\s+', '', regex=True)

# Alternatively, Option 3: Fill with "NOTA" consistently in empty cells for NOTA rows
df_new.loc[df_new['candidate_name'] == 'NOTA', ['candidate_sex', 'candidate_age', 'candidate_category']] = 'NOTA'


# Save the modified DataFrame to a new CSV
df_new.to_csv('final_updated_dataset.csv', index=False)

print("Position column added and CSV exported as 'final_updated_dataset.csv'")
```
