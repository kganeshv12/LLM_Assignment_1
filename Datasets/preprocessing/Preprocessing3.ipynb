```python
import re

def get_abbreviation(party_name):

    if party_name.lower() == "independent":
        return "IND"
    # Define the words to ignore
    ignore_words = {"of", "for", "the", "and", "a", "in", "on", "to", "by", "with"}
    
    # Function to handle words in brackets
    def process_brackets(name):
        # Find all content inside parentheses and process it
        bracketed = re.findall(r'\((.*?)\)', name)
        for b in bracketed:
            # Process the words inside the bracket
            bracket_abbr = ''.join([word[0].upper() for word in b.split() if word.lower() not in ignore_words])
            name = name.replace(f'({b})', f'({bracket_abbr})')
        return name
    
    # Clean up the name by removing any unnecessary spaces and applying bracket transformations
    party_name = process_brackets(party_name)
    
    # Split the name into words and form the abbreviation
    words = party_name.split()
    abbreviation = ''
    for word in words:
        # If the word is inside parentheses, append the abbreviation correctly
        if word.startswith("("):
            abbreviation += " (" + word[1:].upper()
        elif word.endswith(")") and abbreviation.endswith(" "):
            abbreviation = abbreviation.strip() + word.upper()
        # Ignore the specified words and take the first letter of the others
        elif word.lower() not in ignore_words and word != '(' and word != ')':
            abbreviation += word[0].upper()
    
    return abbreviation

# Test cases


# Output the abbreviations
abbreviation_dict = dict()
abbreviation_list = list()
for name in party_names:
    abbreviation_dict[name] = get_abbreviation(name)
    abbreviation_list.append(get_abbreviation(name))
    print(f"{name} -> {get_abbreviation(name)}")
```
